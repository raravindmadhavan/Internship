1.	Scrape  the  details  of  most  viewed  videos  on  YouTube  from  Wikipedia.  Url  = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:  
A) Rank
B) Name   
C) Artist   
D) Upload date   
E) Views   

```python
name=[]
artist=[]
date=[]
views=[]
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos') # Opens wikipedia
driver.maximize_window()
time.sleep(3)
# Gets Name,artist,date,views
name.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[1]")])
artist.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[2]")])
date.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[4]")])
views.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[3]")])
driver.quit()
rank=[i for i in range(1,len(name)+1)]
video_details=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload Date':date,'Views':views})
video_details.head()
```

2) Scrape the details team India’s international fixtures from bcci.tv.   
Url = https://www.bcci.tv/.   
You need to find following details:   
A) Series   
B) Place   
C) Date   
D) Time   
Note: - From bcci.tv home page you have reach to the international fixture page through code. 


```python
tournament=[]
place=[]
date=[]
time_1=[]
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://www.bcci.tv/') # Opens https://www.bcci.tv/.
driver.maximize_window()
d=driver.find_element(By.XPATH,"//*[contains(@data-element_text,'Fixtures')]") # Fixtures tab

d.click() # clicks on Fixtures tab
time.sleep(10)
tournament.extend([i.text for i in driver.find_elements(By.XPATH,"//h5[starts-with(@class,'match-tournament-name')]")])
place.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'match-place')]")])
date.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'match-dates')]")])
time_1.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'match-time')]")])
time.sleep(10)
driver.quit()
match_details=pd.DataFrame({'Tournament':tournament,'Place':place,'Date':date,'Time':time_1})
match_details.head()
```

4. Scrape the details of trending repositories on Github.com.   
Url = https://github.com/   
You have to find the following details:   
A) Repository title   
B) Repository description   
C) Contributors count   
D) Language used 
Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
title=[]
desc=[]
count=[]
lang=[]
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://github.com/') # Opens https://github.com/
driver.maximize_window()
time.sleep(10)
d=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button") # Open Sources
d.click() # clicks on Open Source
e=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a") # `trending
e.click() # clicks on Trending
time.sleep(10)
# Only 10 rows
for j in range(1,11):
    try:
        t=driver.find_element(By.XPATH,f"/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article[{j}]/h2/a")
        title.append(t.text)
    except:
        title.append('-')
    try:
        d=driver.find_element(By.XPATH,f"/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article[{j}]/p")
        desc.append(d.text)
    except:
        desc.append('-')
    try:
        c=driver.find_element(By.XPATH,f"/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article[{j}]/div[2]/a[1]")
        count.append(c.text)
    except:
        count.append('-')
    try:
        l=driver.find_element(By.XPATH,f"/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article[{j}]/div[2]/span[1]/span[2]")
        lang.append(l.text)
    except:
        lang.append('-')
driver.quit()
gitrepo_details=pd.DataFrame({'Title':title,'Description':desc,'Count':count,'Language':lang})
gitrepo_details.head()
```

7. Scrape the details most watched tv series of all time from imdb.com.   
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details:   
A) Name   
B) Year span   
C) Genre   
D) Run time   
E) Ratings   
F) Votes   


```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://www.imdb.com/chart/tvmeter/?ref_=nv_tvv_mptv') # Opens https://www.imdb.com/list/ls095964455/ 
driver.maximize_window()
d=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[1]/div[3]/button") # Open info
d.click() # clicks on info
time.sleep(10)
run_time=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[2]/div/div/div[1]/div[2]/ul[1]/li[2]").text
genre=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[2]/div/div/div[1]/div[2]/ul[2]").text
genre=','.join(re.split('(?<=[a-z])(?=[A-Z])',genre))
ratings=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[2]/div/div/div[1]/div[2]/div[2]/span").text
# Removing spaces in rating
ratings=re.sub('\n','',ratings)
e=driver.find_element(By.XPATH,"/html/body/div[4]/div[2]/div/div[1]/button") # Close
e.click() # clicks on close
movie_name=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[1]/div[2]/div/div/div[2]/a/h3").text
year=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[1]/div[2]/div/div/div[3]/span[1]").text

votes=driver.find_element(By.XPATH,"/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[1]/div[2]/div/div/span/div/span/span").text
# Removing space,(,) characters
votes=re.sub(r'[\s()]','',votes)
driver.quit()
print(movie_name,year,genre,ratings,votes,run_time)
```

 
8. Details of Datasets from UCI machine learning repositories.   
Url = https://archive.ics.uci.edu/  You 
have to find the following details:   
A) Dataset name   
B) Data type   
C) Task   
D) Attribute type   
E) No of instances   
F) No of attribute G) Year   
      Note: - from the home page you have to go to the Show All Dataset page through code.

```python
dataset=[]
datatype=[]
task=[]
attribute_type=[]
instance_count=[]
attribute_count=[]
year=[]
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://archive.ics.uci.edu/') # Opens https://archive.ics.uci.edu/
driver.maximize_window()
time.sleep(10)
d=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]") # View datasets
d.click() # clicks on Charts
time.sleep(10)
e=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]") # Expand all
e.click() # clicks on Expand All
time.sleep(10)
dataset.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div/div[2]/h2/a")])
datatype.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[2]/span")])
task.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[1]/span")])
instance_count.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[3]/span")])
attribute_type.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[2]")])
attribute_count.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[4]/span")])
year.extend([i.text for i in driver.find_elements(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr/td[3]")])
driver.quit()
```

```python
# Extracting the year from the date field
def year_extract(date):
    if date=='N/A':
        return date
    else:
        return date[-4:]
year=list(map(year_extract,year))
year
```

```python
dataset_details=pd.DataFrame({'Dataset':dataset,'Datatype':datatype,'Task':task,'Attribute type':attribute_type,'No of instances':instance_count,'No of attribute':attribute_count,'Year':year})
dataset_details.head()
```
