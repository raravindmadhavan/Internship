3.	Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.  

```python

for search_input in ['Fruits','Cars','Machine Learning','Guitar','Cakes']:
    driver=webdriver.Chrome() # Opening Chrome
    driver.get('https://images.google.com')
    d=driver.find_element(By.XPATH,'//textarea[contains(@class,"gLFyf")]') # Search bar
    d.send_keys(search_input) # Search for images
    search_button=driver.find_element(By.XPATH,'//button[contains(@class,"Tg7LZd")]')  
    search_button.submit() # Click on search
    # Find the images.
    time.sleep(3)
    imgResults = driver.find_elements(By.XPATH,"//img[starts-with(@jsname,'Q4LuWd')]")
    # Access and store the scr list of image url's.
    src = []
    for img in imgResults:
        src.append(img.get_attribute('src'))
    # Retrieve and download the images.
    for i in range(10):    
        urllib.request.urlretrieve(str(src[i]),"{}{}.jpg".format(search_input,i))
    driver.quit()
```

```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://flipkart.com') # Opens flipkart.com
d=driver.find_element(By.CLASS_NAME,'Pke_EE') # Search bar
d.send_keys('Oneplus Nord') # Search for Oneplus Nord
search_button=driver.find_element(By.CLASS_NAME,'_2iLD__')  
search_button.submit() # Click on search
prod_list = driver.find_elements(By.XPATH,"//a[starts-with(@class,'_1fQZEK')]")
brand_name=[]
smartphone_name=[]
colour=[]
ram=[]
storage=[]
cam=[]
size=[]
capacity=[]
price=[]
prod_url=[]
for prod in prod_list:
    brand_name.extend(['OnePlus'])
    smartphone_name.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'_4rR01T')][1]")])
    colour.extend([i.text.split('(')[-1].split(',')[0] for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'_4rR01T')]")])
    ram.extend([i.text.split('|')[0] for i in driver.find_elements(By.XPATH,"(//ul[starts-with(@class,'_1xgFaf')]/li)[1]")])
    storage.extend([i.text.split('|')[-1] for i in driver.find_elements(By.XPATH,"(//ul[starts-with(@class,'_1xgFaf')]/li)[1]")])
    cam.extend([i.text for i in driver.find_elements(By.XPATH,"(//ul[starts-with(@class,'_1xgFaf')]/li)[3]")])
    size.extend([i.text for i in driver.find_elements(By.XPATH,"(//ul[starts-with(@class,'_1xgFaf')]/li)[2]")])
    capacity.extend([i.text for i in driver.find_elements(By.XPATH,"(//ul[starts-with(@class,'_1xgFaf')]/li)[4]")])
    price.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'_30jeq3')]")])
    prod_url.extend([prod.get_attribute('href')])
driver.quit()
prod=pd.DataFrame({'Brand Name':brand_name,'Smart Phone Name':smartphone_name[:24],'Price':price[:24],'Colour':colour[:24],
                   'RAM':ram,
                   'Storage':storage,'Camera':cam,'Size':size,'Capacity':capacity,'Url':prod_url})
prod.shape
prod.head()
```

5.	Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.  

```python
def lat_long(city):
    driver=webdriver.Chrome() # Opening Chrome
    driver.get('https://www.google.com/maps') # https://www.google.com/maps
    d=driver.find_element(By.XPATH,'//input[contains(@id,"searchboxinput")]') # Search bar
    d.send_keys(city) # Search for chennai
    search_button=driver.find_element(By.XPATH,'//button[contains(@aria-label,"Search")]')  
    search_button.click() # Click on search
    time.sleep(5)
    current_url = driver.current_url # current url
    split_url = current_url.split('/@')[1].split(',')
    latitude = split_url[0]
    longitude = split_url[1]
    return latitude,longitude
latitude,longitude = lat_long('Kabul')
driver.quit()
latitude,longitude
```

6.	Write a program to scrap all the available details of best gaming laptops from digit.in. 

```python
def gaming_laptops():
    driver=webdriver.Chrome() # Opening Chrome
    driver.get('https://www.digit.in/top-products/best-gaming-laptops-40.html') # Opens digit.in
    driver.maximize_window()
    laptop_cards = driver.find_elements(By.XPATH,'//div[starts-with(@class,"r_offer_details")]')
    laptops={'Name':[],'OS':[],'Display Size':[],'Resolution':[],'Processor':[]}
    count=0
    for lap in laptop_cards:
        count+=1
        l=lap.text.split('\n')
        laptops['Name'].extend([l[0]])
        if 'Operating System:' in l:
            laptops['OS'].extend([l[2]])
            laptops['Display Size'].extend([l[4]])
            laptops['Resolution'].extend([l[6]])
            laptops['Processor'].extend([l[-1]])
        else:
            laptops['OS'].extend(['-'])
            laptops['Display Size'].extend(['-'])
            laptops['Resolution'].extend(['-'])
            laptops['Processor'].extend(['-'])
    driver.quit()
    return pd.DataFrame(laptops)
gaming_laptops().head()
```

8.	Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.  

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Replace this with the link of the YouTube video you want to scrape
video_url = "https://www.youtube.com/watch?v=-naGPCAxBqk&t=695s"

# Function to scroll down the page to load more comments
def scroll_down(driver):
    # Scroll down the page until no more comments are loaded
    last_height = driver.execute_script("return document.documentElement.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(2)  # Adjust the sleep time as needed
        new_height = driver.execute_script("return document.documentElement.scrollHeight")

        if new_height == last_height:
            break
        last_height = new_height

# Start a Chrome browser session using Selenium
options = webdriver.ChromeOptions()
options.add_argument("--disable-notifications")
driver = webdriver.Chrome(options=options)

# Open the YouTube video URL
driver.get(video_url)
time.sleep(5)  # Allow time for the page to load

# Scroll to load comments
scroll_down(driver)

# Find all comment elements
comments = driver.find_elements(By.XPATH, "//yt-formatted-string[@id='content-text']")

# Find all upvote elements
upvotes = driver.find_elements(By.XPATH, "//span[@id='vote-count-middle']")

# Find all timestamp elements
timestamps = driver.find_elements(By.CSS_SELECTOR, "a.yt-simple-endpoint yt-formatted-string.published-time-text")

# Print the extracted data
for i in range(len(comments)):
    comment = comments[i].text if i < len(comments) else "N/A"
    upvote = upvotes[i].text if i < len(upvotes) else "N/A"
    timestamp = timestamps[i].get_attribute("title") if i < len(timestamps) else "N/A"
    
    print(f"Comment {i+1}: {comment}")
    print(f"Upvotes: {upvote}")
    print(f"Timestamp: {timestamp}")
    print("="*50)

# Close the browser session
driver.quit()
```
