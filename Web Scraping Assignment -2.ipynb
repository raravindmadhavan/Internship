Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
6. Brand
7. ProductDescription
8. Price
The attributes which you have to scrape is ticked marked in the below image.
To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url :https://www.flipkart.com/
2. Enter “sunglasses” in the search fieldwhere “search for products, brands and more” is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses.


```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://flipkart.com') # Opens Shine.com
d=driver.find_element(By.CLASS_NAME,'Pke_EE') # Search bar
d.send_keys('Sunglasses') # Search for sunglasses
search_button=driver.find_element(By.CLASS_NAME,'_2iLD__')  
search_button.submit() # Click on search
# pick 4 parameters of each sunglass
count=0
product_list=[]
desc_list=[]
price_list=[]
offer_list=[]
while count<3:
    try:
        product_list.extend([i.text for i in driver.find_elements(By.CLASS_NAME,'_2WkVRV')]) # Product names
        desc_list.extend([i.text for i in driver.find_elements(By.CLASS_NAME,'IRpwTa')]) # Description
        price_list.extend([i.text for i in driver.find_elements(By.CLASS_NAME,'_30jeq3')]) # Price
        offer_list.extend([i.text for i in driver.find_elements(By.CLASS_NAME,'_3Ay6Sb')]) #Offer
        count+=1
        if count>0: # Clicks on next button
            next_button= driver.find_element(By.CLASS_NAME,"_2MImiq")
            next_button.click()
    except:
        continue
sunglasses_df=pd.DataFrame({'Product':product_list[:100],'Description':desc_list[:100],'Price':price_list[:100],
                                                                                                          'Offer':offer_list[:100]})
sunglasses_df.shape
```

Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market
place=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.
Note: All the steps required during scraping should be done through code only and not manually.

```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market') # Opens flipkart
count=0
rating_list=[]
review_list=[]
fr_list=[]
offer_list=[]
while len(review_list)<100:
    rating_list.extend([i.text for i in driver.find_elements(By.XPATH,"//div[starts-with(@class,'_3LWZlK')]")]) # Rating
    review_list.extend([i.text for i in driver.find_elements(By.XPATH,"//p[contains(@class,'_2-N8zT')]")]) # Review
    fr_list.extend([i.text for i in driver.find_elements(By.XPATH,"//div[contains(@class,'t-ZTKy')]")]) # Price
    # Clicks on next button
    next_button= driver.find_element(By.CLASS_NAME,'_1LKTO3')
    next_button.click()
    print(len(fr_list),len(rating_list),len(review_list))

iphone_review_df=pd.DataFrame({'Rating':rating_list[:100],'Review':review_list[:100],'Full Review':fr_list[:100]})
iphone_review_df.shape
```

Q7 -Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then
set CPU Type filter to “Intel Core i7” as shown in the below image: 
After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price


```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://amazon.in') # Opens amazon.in
d=driver.find_element(By.XPATH,'//input[contains(@id,"twotabsearchtextbox")]') # Search bar
d.send_keys('Laptop') # Search for sneakers
search_button=driver.find_element(By.XPATH,'//input[contains(@id,"nav-search-submit-button")]')  
search_button.submit() # Click on search
set_cpu_type=driver.find_element(By.XPATH,'//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/div/label/i') 
set_cpu_type.click() # Check the Intel Core I7
product_list=[]
rating_list=[]
price_list=[]
product_list.extend([i.text for i in driver.find_elements(By.XPATH,'//span[starts-with(@class,"a-size-medium ")]')]) # Product names
rating_list.extend([i.text for i in driver.find_elements(By.XPATH,'//span[starts-with(@class,"a-icon-alt")]')])
price_list.extend([i.text for i in driver.find_elements(By.XPATH,'//span[starts-with(@class,"a-price-whole")]')])
laptop_df=pd.DataFrame({'Product':product_list[:10],'Rating':rating_list[:10],'Price':price_list[:10]})
laptop_df
```

Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/ 
This task will be done in following steps: 
1. First get the webpage https://www.motor1.com/ 
2. Then You have to type in the search bar ’50 most expensive cars’ 
3. Then click on 50 most expensive cars in the world.. 
4. Then scrap the mentioned data and make the dataframe

```python
driver=webdriver.Chrome() # Opening Chrome
driver.get('https://www.motor1.com/') # Opens https://www.motor1.com/
search_button=driver.find_element(By.XPATH,'//button[contains(@class,"m1-search-form-button-animate")]')  
search_button.click() # Click on search
d=driver.find_element(By.XPATH,'//input[contains(@class,"m1-search-panel-input")]') # Search bar
d.send_keys('50 most expensive cars') # Search for 50 most expensive cars
search_button=driver.find_element(By.XPATH,'//button[contains(@class,"m1-search-panel-button")]')  
search_button.click() # Click on search
search_button=driver.find_element(By.XPATH,'//a[contains(@href,"/features/308149")]')  
search_button.click() # Click on 50 most expensive cars
name_list=[]
price_list=[]
name_list.extend([i.text for i in driver.find_elements(By.XPATH,'//h3[starts-with(@class,"subheader")]')]) # Car
price_list.extend([i.text for i in driver.find_elements(By.XPATH,'//strong')]) # Price
car_price_df=pd.DataFrame({'Car':name_list[:50],'Price':price_list})
car_price_df
```
