1)	Write a python program to display all the header tags from wikipedia.org and make data frame. 

```python
page=requests.get('http://www.wikipedia.org')
soup=bs(page.content)
# Extracting header tags
header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
# Removing extra spaces
header_text = [tag.text.strip() for tag in header_tags]
# Creating dataframe
df = pd.DataFrame({'Header Tags': header_text})
df
```

3)	Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame- 
a)	Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 
b)	Top 10 ODI Batsmen along with the records of their team andrating. 
c)	Top 10 ODI bowlers along with the records of their team andrating. 

```python
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
men_rank=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
teams=[]
matches=[]
points=[]
ratings=[]
table = bs(men_rank.content).find('table', class_='table')
table_body=table.find('tbody')
rows=table_body.find_all('tr')
for row in rows[:10]:  # Considering only the top 10 teams
    cols = row.find_all('td')
    team = cols[1].find('span',class_='u-hide-phablet').text.strip()
    teams.append(team)

    match = cols[2].text.strip()
    matches.append(match)

    point = cols[3].text.strip()
    points.append(point)

    rating = cols[4].text.strip()
    ratings.append(rating)
# Creating a DataFrame
data = {
    'Team': teams,
    'Matches': matches,
    'Points': points,
    'Rating': ratings
}

men_rank = pd.DataFrame(data)

men_rank
```

```python
#b) Top 10 ODI Batsmen along with the records of their team andrating.
bat_rank=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
bat_name=[]
team_name=[]
bat_ratings=[]
bat_rank_soup=bs(bat_rank.content)
table_1 = bat_rank_soup.find('table', class_='table')
table_body_1=table_1.find('tbody')
rows_1=table_body_1.find_all('tr')
bat_name.append(bat_rank_soup.find('div', class_='rankings-block__banner--name').text.strip())
team_name.append(bat_rank_soup.find('div', class_='rankings-block__banner--nationality').text.strip().split('\n')[0])
bat_ratings.append(bat_rank_soup.find('div', class_='rankings-block__banner--rating').text.strip())
for row in rows_1[0:9]:  # 2 to 10 batsman
    cols = row.find_all('td')
    bat_name.append(cols[1].text.strip())
    team_name.append(cols[2].find('span',class_="table-body__logo-text").text)
    bat_ratings.append(cols[3].text)
# Creating a DataFrame
data_1 = {
    'Batsman': bat_name,
    'Team': team_name,
    'Rating': bat_ratings
}
batsmen_rank = pd.DataFrame(data_1)
batsmen_rank
```

```python
# c) Top 10 ODI bowlers along with the records of their team andrating.
bowl_rank=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
bowl_name=[]
team_name=[]
bowl_ratings=[]
bowl_rank_soup=bs(bowl_rank.content)
table_2 = bowl_rank_soup.find_all('table', class_='table')[1]
table_body_2=table_2.find('tbody')
rows_2=table_body_2.find_all('tr')
bowl_name.append(bowl_rank_soup.find_all('div', class_='rankings-block__banner--name')[1].text.strip())
team_name.append(bowl_rank_soup.find_all('div', class_='rankings-block__banner--nationality')[1].text.strip().split('\n')[0])
bowl_ratings.append(bowl_rank_soup.find_all('div', class_='rankings-block__banner--rating')[1].text.strip())
for row in rows_2[0:9]:  # 2 to 10 batsman
    cols = row.find_all('td')
    bowl_name.append(cols[1].text.strip())
    team_name.append(cols[2].find('span',class_="table-body__logo-text").text)
    bowl_ratings.append(cols[3].text)
# Creating a DataFrame
data_2 = {
    'Bowler': bowl_name,
    'Team': team_name,
    'Rating': bowl_ratings
}
bowler_rank = pd.DataFrame(data_2)
bowler_rank
```

5)	Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame- 
i) Headline 
ii) Time 
iii) News Link 

```python
url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

headlines = []
times = []
news_links = []

articles = soup.find_all('div', class_='LatestNews-container')
for article in articles:
    headline = article.find('a').text.strip()

    time = article.find('time').text.strip()

    news_link =article.find('a')['href']
    if news_link.startswith('/'):
        continue
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)

# Creating a DataFrame
data = {
    'Headline': headlines,
    'Time': times,
    'News Link': news_links
}

df = pd.DataFrame(data)
df.dropna(inplace=True)
df
```

Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles 
Scrape below mentioned details and make data frame- 
i)	Paper Title 
ii)	Authors 
iii)	Published Date 
iv)	Paper URL 

```python
url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

titles = []
authors = []
published_dates = []
paper_urls = []

articles = soup.find_all('li',class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs")
for article in articles:
    title = article.find('h2').text.strip()
    titles.append(title)
    author = article.find('p').text.strip()
    authors.append(author)
    published_date = article.find('span', class_='sc-1thf9ly-2 dvggWt').text.strip()
    published_dates.append(published_date)
    paper_url = article.find('a')['href']
    paper_urls.append(paper_url)

# Creating a DataFrame
data_4 = {
    'Paper Title': titles,
    'Authors': authors,
    'Published Date': published_dates,
    'Paper URL': paper_urls
}

journals= pd.DataFrame(data_4)
journals
```
